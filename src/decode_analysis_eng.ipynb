```python
def plot_multi(df, filters, gte=True, max_x=None, lookup='f1s', line_at_quantiles=None, line_at_lengths=None, xlim=None, ylim=None):
    fig, ax = plt.subplots()
    colors = ['r', 'g', 'b', 'y']
    ix = 0
    quantile_lengths = None
    for name, filt in filters.items():
        df_filt = df[filt]
        span_counts = df_filt['span_counts']
        all_stats = span_counts.apply(get_span_f1s_gte_from_counts if gte else get_span_f1s_from_counts)
        acc = all_stats.apply(
            lambda stats: pandas.Series(stats._asdict()[lookup], stats.lengths)
        ).mean(axis=0)
        if max_x:
            acc = acc[acc.index <= max_x]
        acc.plot(ax=ax, label=name)
        if line_at_quantiles:
            quant_lens = []
            for quantile in line_at_quantiles:
                gold_counts = span_counts.iloc[0].gold_counts
                gold_counts_cum = np.cumsum(gold_counts)
                ins_point = bisect.bisect_right(gold_counts_cum, gold_counts_cum[-1] * quantile)
                print("{} / {}".format(gold_counts_cum[ins_point], gold_counts_cum[-1]))
                quant_lens.append(span_counts.iloc[0].lengths[ins_point])
                print("quan")
            print("quant_lens: {}".format(quant_lens))
            print("quant_lens: {}".format(quant_lens))
            if quantile_lengths is not None:
                assert quantile_lengths == quant_lens
            else:
                quantile_lengths = quant_lens
    if xlim:
        ax.set_xlim(xlim)
    if ylim:
        ax.set_ylim(ylim)
    if quantile_lengths:
        ax.vlines(quantile_lengths, *ax.get_ylim())
    if line_at_lengths:
        ax.vlines(line_at_lengths, *ax.get_ylim())
    ax.legend()
```

```python
def plot_accuracies(df, groupby, gte=True):
    fig, ax = plt.subplots()
    colors = ['r', 'g', 'b', 'y']
    ix = 0
    lookup = 'span_accuracies_gte_xy' if gte else 'span_accuracies_xy'
    for key, group in df.groupby(groupby):
        assert(len(group) == 1)
        ax.plot(*group[lookup][0], label=' '.join(key))
    ax.legend()
```
