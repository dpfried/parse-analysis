```python
def show_splits(df, key='fscore', split='vocab'):
    def get(part):
        return df[df.index.get_level_values('corpus_name') == part].iloc[0][key]
    print("{0} dis \t {1:.2f}".format(split, get('wsj_test_{}_dis'.format(split))))
    print("{0} sim \t {1:.2f}".format(split, get('wsj_test_{}_sim'.format(split))))
```

```python
def show_quadrants(df, key='fscore'):
    def get(part):
        return df[df.index.get_level_values('corpus_name') == part].iloc[0][key]
    print("\t g0 \t g1")
    print("v0 \t {0:.2f} \t {1:.2f}".format(get('wsj_test_v0g0'), get('wsj_test_v0g1')))
    print("v1 \t {0:.2f} \t {1:.2f}".format(get('wsj_test_v1g0'), get('wsj_test_v1g1')))
```

```python
for name in ['inorder', 'chart', 'charniak-WSJ-PTB3']:
    print(name)
    show_splits(df[df.index.get_level_values('parser') == name])
    print()
```

```python
for name in ['inorder', 'chart', 'charniak-WSJ-PTB3']:
    print(name)
    show_splits(df[df.index.get_level_values('parser') == name], split='mwp')
    print()
```
